```python
import pandas as pd
import os
```

**imports.csv and exports.csv are files downloaded from https://oec.world/en/visualize/tree_map/hs92/export/usa/show/all/2018.2019.2020/. The code below merges them and downloads this merged dataframe to a file called "ustrade.csv"**

```python
imports = pd.read_csv("imports.csv")
imports = imports.rename(columns={"Trade Value": "Imports"})
imports = imports.drop(columns="ISO 3")
exports = pd.read_csv("exports.csv")
exports = exports.rename(columns={"Trade Value": "Exports"})
exports = exports.drop(columns="ISO 3")
```

```python
imports.head()
```

```python
exports.head()
```

```python
combined = imports.merge(exports, on=["Year", "Country", "Continent", "Continent ID", "Country ID"])
combined = combined[["Year", "Country", "Continent", "Imports", "Exports"]]
combined.head()
```

```python
combined["Deficit/Surplus"] = combined["Exports"] - combined["Imports"]
combined.head()
```

```python
combined.groupby("Year").agg({"Deficit/Surplus": sum})
```

```python
filename = "ustrade.csv"
if not os.path.exists(filename):
    combined.to_csv(filename, index=False)
else:
    print(f"The file named {filename} already exists in this folder")
```
